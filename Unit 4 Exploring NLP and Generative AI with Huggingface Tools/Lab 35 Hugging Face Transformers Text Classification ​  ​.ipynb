```python
from transformers import pipeline
```

#### Step 3: Perform Text Classification
Use the pipeline feature in Transformers for simple text classification:

```python
# Example text for sentiment analysis
text = "I didn't like the movie, it was disappointing."

# Perform sentiment analysis using a pretrained model
classifier = pipeline('sentiment-analysis')
result = classifier(text)

# Extract predicted label and score
predicted_label = result[0]['label']
confidence_score = result[0]['score']

# Check if the predicted sentiment is negative
if predicted_label == 'NEGATIVE':
    print(f"The sentiment analysis indicates a negative sentiment for the text: '{text}'")
    print(f"Confidence score: {confidence_score:.4f}")
else:
    print(f"The sentiment analysis indicates a positive sentiment for the text: '{text}'")
    print(f"Confidence score: {confidence_score:.4f}")
```
