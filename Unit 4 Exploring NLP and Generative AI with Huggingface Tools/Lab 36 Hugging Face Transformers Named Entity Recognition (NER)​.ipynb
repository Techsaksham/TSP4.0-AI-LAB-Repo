#### Step 2: Import Libraries
Import the required libraries

```python
from transformers import pipeline
```

#### Step 3: Load Pretrained NER Model
Load a pretrained NER model from Hugging Face Transformers

```python
# Load NER pipeline
ner = pipeline("ner")
```

#### Step 4: Perform Named Entity Recognition
Use the loaded NER pipeline to identify entities in a text

```python
# Example text for NER
text = "Apple Inc. is an American multinational technology company headquartered in Cupertino, California."

# Perform NER on the text
ner_results = ner(text)

# Output the identified entities
for entity in ner_results:
    print(f"Entity: {entity['word']}, Type: {entity['entity']}")
```

This example demonstrates a simple approach to perform Named Entity Recognition (NER) using Hugging Face Transformers. The pretrained model handles the task of identifying entities in text efficiently. Adjust the input text variable with different sentences to observe how the NER model identifies various named entities and their types.
