## Import necessary libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Helper function to display images in Jupyter Notebook
def display_image(img, title="Image", cmap_type='gray'):
    plt.figure(figsize=(10, 8))
    plt.imshow(img, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

### 1. Image Read and Display by Matplotlib

```python

# Read an image using OpenCV
img = cv2.imread('Couple.jpg')  # replace 'your_image_path.jpg' with your image file path

# Convert the image from BGR to RGB format
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using Matplotlib
display_image(img_rgb, title="Original Image", cmap_type=None)
```

```python
# 2. Resize Image and Display
# Resize the image
img_resized = cv2.resize(img_rgb, (200, 200))

# Display the resized image
display_image(img_resized, title="Resized Image", cmap_type=None)
```

```python
# 3. Image Annotations
# Draw a rectangle on the image
img_annotated = img_rgb.copy()
cv2.rectangle(img_annotated, (50, 50), (150, 150), (255, 0, 0), 3)  # BGR color format, here it is blue

# Draw a circle on the image
cv2.circle(img_annotated, (100, 100), 30, (0, 255, 0), 3)  # BGR color format, here it is green

# Add text to the image
cv2.putText(img_annotated, 'Hello', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)  # BGR color format, here it is red

# Display the annotated image
display_image(img_annotated, title="Annotated Image", cmap_type=None)
```

```python
# 4. Image Thresholding
# Convert the image to grayscale
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

# Apply binary thresholding
_, img_thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

# Display the thresholded image
display_image(img_thresh, title="Thresholded Image")
```

```python
# 5. Edge Detection
# Apply Canny edge detection
edges = cv2.Canny(img_gray, 100, 200)

# Display the edges
display_image(edges, title="Edge Detection")
```
